```python
#Get risk-free rate
rf = db.raw_sql("select mcaldt, tmytm from crsp.tfz_mth_rf where kytreasnox=2000001 and mcaldt>='1978-01-01' and mcaldt<='2020-12-31'")
rf.to_csv('rf.csv')
```

```python
rf = pd.read_csv('rf.csv', usecols=['tmytm']).squeeze()
#Compute the monthly rate
rf_m = (np.exp(rf/100)-1)/12
```

```python
#Apply ten year window
ret_1 = ret.loc['19780101':'19880101']
ret_2 = ret.loc['19880101':'19980101']
ret_3 = ret.loc['19980101':'20080101']
ret_4 = ret.loc['20080101':'20180101']

rf_m = np.array(rf_m)

#Get the mean and std of return for each stock during the ten year window
mean_1 = ret_1.mean()
std_1 = ret_1.std()
cov_1 = np.cov(np.matrix.transpose(np.array(ret_1)))
rf_1 = rf_m[0:120].mean()

mean_2 = ret_2.mean()
std_2 = ret_2.std()
cov_2 = np.cov(np.matrix.transpose(np.array(ret_2)))
rf_2 = rf_m[121:240].mean()

mean_3 = ret_3.mean()
std_3 = ret_3.std()
cov_3 = np.cov(np.matrix.transpose(np.array(ret_3)))
rf_3 = rf_m[241:360].mean()

mean_4 = ret_4.mean()
std_4 = ret_4.std()
cov_4 = np.cov(np.matrix.transpose(np.array(ret_4)))
rf_4 = rf_m[361:480].mean()
```

### a)

```python
#equally weighted portfolio
rp1 = mean_1.mean()
stdp1 = np.sqrt(cov_1.sum()/(20*20))
sh1 = (rp1-rf_1)/stdp1

rp2 = mean_2.mean()
stdp2 = np.sqrt(cov_2.sum()/(20*20))
sh2 = (rp2-rf_2)/stdp2

rp3 = mean_3.mean()
stdp3 = np.sqrt(cov_3.sum()/(20*20))
sh3 = (rp3-rf_3)/stdp3

rp4 = mean_4.mean()
stdp4 = np.sqrt(cov_4.sum()/(20*20))
sh4 = (rp4-rf_4)/stdp4

rp_equal_annual = (rp1+rp2+rp3+rp4)/4*12
std_equal_annual = np.sqrt(stdp1**2 + stdp2**2 + stdp3 **2 + stdp4**2)*np.sqrt(12)
sh_equal_annual = (rp_equal_annual-(rf_1+rf_2+rf_3+rf_4)/4*12)/std_equal_annual

print("The monthly portfolio return at the beginning of 1988 is ", rp1, "std is ", stdp1, "Sharpe ratio is ", sh1, "\n")
print("The monthly portfolio return at the beginning of 1998 is ", rp2, "std is ", stdp2, "Sharpe ratio is ", sh2, "\n")
print("The monthly portfolio return at the beginning of 2008 is ", rp3, "std is ", stdp3, "Sharpe ratio is ", sh3, "\n")
print("The monthly portfolio return at the beginning of 2018 is ", rp4, "std is ", stdp4, "Sharpe ratio is ", sh4, "\n")
print("The annual mean portfolio return from 1978~2018 is ", rp_equal_annual, "std is ", std_equal_annual, "Sharpe ratio is ", sh_equal_annual, "\n")
```
