## 2. Market capitalization and expected returns

```python
import wrds
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
db = wrds.Connection()
db.create_pgpass_file()
```

```python
# get rf
rf=db.raw_sql("select mcaldt, tmytm from crsp.tfz_mth_rf where kytreasnox = 2000001 and mcaldt>='1980-01-01' and mcaldt<='2019-12-31'")
rf.to_csv('rf.csv')
# get value-weighted CRSP market return
vwret = db.raw_sql("select date, vwretd from crsp.msi where date>='1980-01-01' and date<='2019-12-31'")
vwret.to_csv('vwret.csv')
```

```python
rf = pd.read_csv('rf.csv', usecols=['tmytm']).squeeze()
#Compute the monthly simple rate
rf_m = np.exp(rf/12/100)-1

vwret = pd.read_csv('vwret.csv', usecols=['vwretd']).squeeze()
```

```python
stocks = db.raw_sql("""
                select a.permno, a.date, a.ret, b.exchcd, b.shrcd, a.shrout, a.prc 

                from crsp.msf as a left join crsp.msenames as b 
                
                on a.permno = b.permno and b.namedt <= a.date and a.date<=b.nameendt
                
                where a.date between '1980-01-01' and '2019-12-31'
                and b.exchcd between 1 and 2
                and b.shrcd between 10 and 11""",date_cols=['date'])

stocks.to_csv('stocks.csv')
```

```python
# change table format from long to wide
stocks_ret = pd.pivot_table(data=stocks, values='ret', index='date', columns='permno')
stocks_shrout = pd.pivot_table(data=stocks, values='shrout', index='date', columns='permno')
stocks_prc = pd.pivot_table(data=stocks, values='prc', index='date', columns='permno')
```

```python
s = stocks_ret.apply(sum)
s = list(s.isna())
s_i = []
for i in range(len(s)):
    if not s[i]:
        s_i.append(i)
```

```python
# removing stocks containing NaN values
stocks_ret = stocks_ret.iloc[:,s_i]
stocks_shrout = stocks_shrout.iloc[:,s_i]
stocks_prc = stocks_prc.iloc[:,s_i]
```

### b)

```python
stocks_prc = stocks_prc.abs()
```

```python
for j in range(stocks_shrout.shape[1]):
    for i in range(stocks_shrout.shape[0]):
        if pd.isna(stocks_shrout.iloc[i,j]) or pd.isnull(stocks_shrout.iloc[i,j]):
            if i == 0:
                s = 1
                while pd.isna(stocks_shrout.iloc[i+s,j]) or pd.isnull(stocks_shrout.iloc[i+s,j]):
                    s += 1
                stocks_shrout.iloc[i:(i+s+1),j] = stocks_shrout.iloc[i+s,j]
            else:
                stocks_shrout.iloc[i,j] = stocks_shrout.iloc[i-1,j]
```

```python
for j in range(stocks_prc.shape[1]):
    for i in range(stocks_prc.shape[0]):
        if pd.isna(stocks_prc.iloc[i,j]) or pd.isnull(stocks_prc.iloc[i,j]):
            if i == 0:
                s = 1
                while pd.isna(stocks_prc.iloc[i+s,j]) or pd.isnull(stocks_prc.iloc[i+s,j]):
                    s += 1
                stocks_prc.iloc[i:(i+s+1),j] = stocks_prc.iloc[i+s,j]
            else:
                stocks_prc.iloc[i,j] = stocks_prc.iloc[i-1,j]
```

```python
stocks_mcap = pd.DataFrame(stocks_shrout.values*stocks_prc.values, columns=stocks_shrout.columns, index=stocks_shrout.index)
```

```python
last_mcap = stocks_mcap.iloc[len(stocks_mcap)-1,:]
permnos = np.array(last_mcap.index)
mcap_ord = np.array(np.argsort(last_mcap))
mcap_sort = np.sort(last_mcap)
```

```python
len(mcap_ord)
```

```python
#take the last one out, and put in the last group, each portfolio has 27 stocks except the last one
last = mcap_sort[-1]
ord_last = mcap_ord[-1]
mcap_groups = np.reshape(mcap_sort[:-1],(10,27))
ord_groups = np.reshape(mcap_ord[:-1],(10,27))
```

```python
vwret_e = vwret - rf_m
stocks_e = stocks_ret - rf_m
rf_m.index = stocks_ret.index
stocks_e = stocks_ret.subtract(rf_m, axis = 0)
```

```python
#Equally weighted portfolio

alphas = []
betas = []
x = np.array(vwret_e).reshape((-1,1))

#mean excess return
avg_e = np.zeros(10)

for i in range(9):
    #equally weighted monthly return
    eq_ret = stocks_ret.iloc[:,ord_groups[i]].mean(axis=1)
    
    #equally weighted excess return
    eq_ret_e = eq_ret.subtract(rf_m, axis = 0)
    avg_e[i] = eq_ret_e.mean()
    
    #regression
    y = np.array(eq_ret_e).reshape((-1,1))
    lm = LinearRegression().fit(x,y)
    betas.append(lm.coef_.item())
    alphas.append(lm.intercept_.item()) 
         
    
group10 = np.append(ord_groups[9],ord_last)
eq_ret10 = stocks_ret.iloc[:,group10].mean(axis=1)    
eq_ret_e10 = eq_ret10.subtract(rf_m, axis = 0)
avg_e[9] = eq_ret_e10.mean()
y = np.array(eq_ret_e10).reshape((-1,1))
lm = LinearRegression().fit(x,y)
betas.append(lm.coef_.item())
alphas.append(lm.intercept_.item())

```

```python
m1,b1 = np.polyfit(betas, avg_e, 1)
plt.figure()
plt.scatter(betas, avg_e)
plt.scatter(1, np.mean(vwret))
plt.plot(betas, b1+m1*np.array(betas), label='regression line')
plt.plot(betas, vwret.mean()*np.array(betas), label='CAPM line')
plt.legend()
plt.xlabel('Beta')
plt.ylabel('Average monthly return')
plt.title('Equally weighted portfolio');
```

```python
#Value weighted 

al = []
be = []
x = np.array(vwret_e).reshape((-1,1))
w = np.zeros((stocks_mcap.shape[0],27))
weight = np.zeros((stocks_mcap.shape[0],27))

#mean excess return
avg_vw_e = np.zeros(10)

for i in range(9):
    #value weighted monthly return
    stocks_mcap.iloc[:,ord_groups[i]]
    for j in range(stocks_mcap.shape[0]):
        ww = stocks_mcap.iloc[j,ord_groups[i]]
        w[j,:] =np.reshape(np.array(ww)/ww.sum(),(1,27))
    weight[1:,:] = w[:-1,:]
    vw_r = stocks_ret.iloc[:,ord_groups[i]] * weight
    vw_ret = vw_r.sum(axis = 1)
    
    #value weighted excess return
    vw_ret_e = vw_ret.subtract(rf_m, axis = 0)
    avg_vw_e[i] = vw_ret_e.mean()
    
    #regression
    y = np.array(vw_ret_e).reshape((-1,1))
    lm = LinearRegression().fit(x,y)
    be.append(lm.coef_.item())
    al.append(lm.intercept_.item()) 
         
    
gr10 = np.append(ord_groups[9],ord_last)
for k in range(stocks_mcap.shape[0]):
        ww = stocks_mcap.iloc[j,ord_groups[9]]
        w[k,:] =np.reshape(np.array(ww)/ww.sum(),(1,27))
        weight[1:,:] = w[:-1,:]
vw_r10 = stocks_ret.iloc[:,ord_groups[9]] * weight
vw_ret10 = vw_r10.sum(axis = 1)
vw_ret_e10 = vw_ret10 - rf_m
avg_vw_e[9] = vw_ret_e10.mean()
y = np.array(vw_ret_e10).reshape((-1,1))
lm = LinearRegression().fit(x,y)
be.append(lm.coef_.item())
al.append(lm.intercept_.item())


```

```python
m1,b1 = np.polyfit(be, avg_vw_e, 1)
plt.figure()
plt.scatter(be, avg_vw_e)
plt.scatter(1, np.mean(vwret))
plt.plot(be, b1+m1*np.array(be), label='regression line')
plt.plot(be, vwret.mean()*np.array(be), label='CAPM line')
plt.legend()
plt.xlabel('Beta')
plt.ylabel('Average monthly return')
plt.title('Value weighted portfolio');
```

```python
plt.figure()
plt.plot(alphas, label='equally weighted')
plt.plot(al, label='value weighted')
plt.legend()
plt.title('Alphas')
plt.xlabel('Decile portfolio')


plt.figure()
plt.plot(betas, label='equally weighted')
plt.plot(be, label='value weighted')
plt.legend()
plt.title('Betas')
plt.xlabel('Decile portfolio')

plt.figure()
plt.plot(avg_e, label='equally weighted')
plt.plot(avg_vw_e, label='value weighted')
plt.legend()
plt.title('Mean excess return')
plt.xlabel('Decile portfolio')
```

### c)

From what we have in part b), market capitalization does not seem to have a relation with average excess returns.
For equally weighted portfolio, as we shown in the plot the fitted regression line and the CAPM line, it does not seem to be a good fit. For the value weighted portfolio, the CAPM line is almost overlapping the fitted line. Thus, we conclude that CAPM generally holds for the value weighted portfolio.

```python
p_ret = np.zeros((480,10))
p_ret_vw = np.zeros((480,10))
for t in range(stocks_mcap.shape[0]):
    if t > 0:
    #for easiness of calculation disregard the last stock 
        mcap_order = np.array(np.argsort(stocks_mcap.iloc[t-1,:]))
        index = np.reshape(mcap_order[:-1],(10,27))
        for i in range(10):
            #equal weighted portfolio return
            p_ret[t,i] = stocks_ret.iloc[t,index[i]].mean()
    
            #value weighted portfolio return
            ww = stocks_mcap.iloc[t,index[i]]
            w = np.array(ww/ww.sum())
            p_ret_vw[t,i] = np.sum(stocks_ret.iloc[t,index[i]] * w)
    
p_ret = p_ret[1:,:]
p_ret_vw = p_ret_vw[1:,:]
    
```

```python
#equally weighted
alphas = []
betas = []
x = np.array(vwret_e[1:]).reshape((-1,1))

#mean excess return
avg_e = p_ret.mean(axis=0)

for i in range(10):
    
    #regression
    y = np.array(p_ret[:,i]).reshape((-1,1))
    lm = LinearRegression().fit(x,y)
    betas.append(lm.coef_.item())
    alphas.append(lm.intercept_.item()) 

m1,b1 = np.polyfit(betas, avg_e, 1)
plt.figure()
plt.scatter(betas, avg_e)
plt.scatter(1, np.mean(vwret))
plt.plot(betas, b1+m1*np.array(betas), label='regression line')
plt.plot(betas, vwret.mean()*np.array(betas), label='CAPM line')
plt.legend()
plt.xlabel('Beta')
plt.ylabel('Average monthly return')
plt.title('Equally weighted portfolio');
```

```python
#value weighted
al = []
be = []
x = np.array(vwret_e[1:]).reshape((-1,1))

#mean excess return
avg_vw_e = p_ret_vw.mean(axis=0)

for i in range(10):
    
    #regression
    y = np.array(p_ret_vw[:,i]).reshape((-1,1))
    lm = LinearRegression().fit(x,y)
    be.append(lm.coef_.item())
    al.append(lm.intercept_.item()) 
    
m1,b1 = np.polyfit(be, avg_vw_e, 1)
plt.figure()
plt.scatter(be, avg_vw_e)
plt.scatter(1, np.mean(vwret))
plt.plot(be, b1+m1*np.array(be), label='regression line')
plt.plot(be, vwret.mean()*np.array(be), label='CAPM line')
plt.legend()
plt.xlabel('Beta')
plt.ylabel('Average monthly return')
plt.title('Value weighted portfolio');
```

```python
plt.figure()
plt.plot(alphas, label='equally weighted')
plt.plot(al, label='value weighted')
plt.legend()
plt.title('Alphas')
plt.xlabel('Decile portfolio')


plt.figure()
plt.plot(betas, label='equally weighted')
plt.plot(be, label='value weighted')
plt.legend()
plt.title('Betas')
plt.xlabel('Decile portfolio')

plt.figure()
plt.plot(avg_e, label='equally weighted')
plt.plot(avg_vw_e, label='value weighted')
plt.legend()
plt.title('Mean excess return')
plt.xlabel('Decile portfolio')
```

It seems that larger market capitalization leads to smaller mean excess return, with one exception at the 3rd one.

As we can see in the plot of the fitted regression line and the CAPM line, without looking ahead in the data, the two lines are obviously not overlapping for both equally weighted portfolio and the value weighted portfolio.

In the separate comparison plot of the alphas, betas and the mena excess returns, for the two weighting methods, the results are really similar, especially the betas. Compared to in part b, where betas are so different.

Thus, without looking ahead, CAPM cannot explain the behavior of the stocks. Without looking ahead, the behavior of equally weighted portfolio and the value weighted portfolio are similar.
